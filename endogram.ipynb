```python
# Import the necessary libraries
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Use inline plotting for Jupyter
%matplotlib inline

# Datos de los puntos
data = [[18], [22], [25], [27], [42], [43]]

# Generar el linkage con el método "single" que utiliza la mínima distancia (Min Distance)
Z = linkage(data, method='single')

# Graficar el dendrograma
plt.figure(figsize=(10, 6))
dendrogram(Z, labels=[str(x[0]) for x in data])
plt.title("Dendrograma usando Min Distance (Single Linkage)")
plt.xlabel("Puntos de Datos")
plt.ylabel("Altura")
plt.show()
```
